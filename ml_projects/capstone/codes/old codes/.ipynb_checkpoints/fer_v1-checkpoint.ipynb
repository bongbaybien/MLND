# Read in data, format, & visualize

```python
emotion_dict = {0:'angry', 1:'disgust', 2:'fear', 3:'happy', 4:'sad', 5:'surprise', 6:'neutral'}
image_shape = (48,48,1)
n_classes = len(emotion_dict)
n_features = np.product(image_shape)

# read in data
file_path = './data/Kaggle-FER/fer2013.csv'
data_raw = helper.read_in_data(file_path) # data is a tuple of (x_list, y_list, usage_list)

# remove blanks, over sample, convert data to numpy array
data_formatted = helper.format(data_raw, image_shape, n_classes) # return tuple of (x_array, y_array, usage_list)

# number of example for each class
label, count = np.unique(data_formatted[1], return_counts=True)
dict(zip(label, count))

# split into train & test sets. Test set = public, private set is used for final testing, to be consistent with the competition
(x_train, y_train), (x_test, y_test), (x_private, y_private) = helper.create_train_test(data_formatted, n_classes)

# one-hot encode
y_train_onehot = keras.utils.to_categorical(y_train, n_classes)
y_test_onehot = keras.utils.to_categorical(y_test, n_classes)
y_private_onehot = keras.utils.to_categorical(y_private, n_classes)
```

# Build model & train

```python
########## arguments ##########
batch_size = 256
n_ensembles = 5
l2_alpha = 0.001

kernel_regularizer = regularizers.l2(l2_alpha)

cp_name = './models/'

########## ensembles ##########
histories = []
for i in range(n_ensembles):
    # build model
    model_name = 'model_' + str(i) +'.hdf5'
    model = helper.build_model(model_name, image_shape, n_classes
                               , kernel_regularizer=kernel_regularizer)

    # preprocess data, compile & train model
    history = helper.train(x_train, y_train_onehot, x_test, y_test_onehot, x_private, y_private_onehot
                           , batch_size, model, cp_name+model_name)
    histories.append(history)
```

```python
########## save histories ##########
for i, history in enumerate(histories):
    pickle.dump(history.history, open('history_' + str(i) + '.p', 'wb'))
```

```python
########## visualize metrics ##########
for i, history in enumerate(histories):
    plt.figure(figsize=(16,4))
    plt.subplot(len(histories), 1, i+1)
    helper.visualize_metrics(history)
    plt.tight_layout()
```

```python
########## predict & accuracy ##########

train_datagen, test_datagen = helper.preprocess(x_train)
n_test = len(x_private)

# predict for each model
y_preds = []
for i in range(n_ensembles):
    # load best model
    model_name = 'model_' + str(i) + '.hdf5'
    model = helper.build_model(model_name, image_shape, n_classes
                               , kernel_regularizer=kernel_regularizer)
    model.load_weights(cp_name + model_name)
    
    # predict for each model
    y_pred = model.predict_generator(test_datagen.flow(x_private, batch_size=batch_size, shuffle=False)
                                               , steps=n_test//batch_size+1)
    # accuracy for each model
    accuracy = metrics.accuracy_score(y_private.reshape(-1), np.argmax(y_pred, axis=1)) * 100
    print('accuracy for model', i, ':', accuracy, '\n')
    
    # add to ensemble
    y_preds.append(y_pred)

# average the results of all models
y_preds = np.asarray(y_preds)
y_preds_mean = np.argmax(np.mean(y_preds, axis=0), axis=1)
accuracy = metrics.accuracy_score(y_private.reshape(-1), y_preds_mean) * 100
print('accuracy for ensemble :', accuracy, '\n')
```

```python
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers.convolutional import Cropping2D, Conv2D, ZeroPadding2D
from keras.layers.pooling import MaxPooling2D, AveragePooling2D
from keras.layers.core import Flatten, Dropout, Dense
from keras import optimizers
from keras.callbacks import ReduceLROnPlateau, ModelCheckpoint, EarlyStopping, TensorBoard
import keras.backend as k
from keras.initializers import TruncatedNormal
from keras.layers.normalization import BatchNormalization
from keras.layers import Activation
from keras.layers.advanced_activations import LeakyReLU
import keras
kernel_initializer=TruncatedNormal()

model_name = 'test'
model2 = Sequential(name=model_name)

model2.add(Cropping2D(cropping=3, input_shape=image_shape, name='block1_crop'))
model2.add(Conv2D(32, (5,5), padding='same', kernel_initializer=kernel_initializer
            , kernel_regularizer=kernel_regularizer
            , name='block1_conv'
            , weights=model.get_layer('block1_conv').get_weights()))
model2.add(BatchNormalization(axis=3, name='block1_bn'
                            , weights=model.get_layer('block1_bn').get_weights()))
model2.add(LeakyReLU(name='block1_relu'))
model2.add(MaxPooling2D((2,2), strides=(2,2), padding='valid', name='block1_pool'))


model2.add(ZeroPadding2D(padding=1, name='block2_pad'))
model2.add(Conv2D(32, (4,4), padding='valid', kernel_initializer=kernel_initializer
            , kernel_regularizer=kernel_regularizer
            , name='block2_conv'
            , weights=model.get_layer('block2_conv').get_weights()))
model2.add(BatchNormalization(axis=3, name='block2_bn'
                            , weights=model.get_layer('block2_bn').get_weights()))
model2.add(LeakyReLU(name='block2_relu'))
model2.add(AveragePooling2D((2,2), strides=(2,2), padding='valid', name='block2_pool'))

model2.add(Conv2D(64, (5,5), padding='same', kernel_initializer=kernel_initializer
            , kernel_regularizer=kernel_regularizer
            , name='block3_conv'
            , weights=model.get_layer('block3_conv').get_weights()))
model2.add(BatchNormalization(axis=3, name='block3_bn'
                              , weights=model.get_layer('block3_bn').get_weights()))
model2.add(LeakyReLU(name='block3_relu'))
model2.add(AveragePooling2D((2,2), strides=(2,2), padding='valid', name='block3_pool'))

model2.add(Flatten(name='block4_flat'))
model2.add(Dropout(0.2, name='block4_dropout'))
model2.add(Dense(3072, kernel_initializer=kernel_initializer, kernel_regularizer=kernel_regularizer, name='block4_dense'
                , weights=model.get_layer('block4_fc').get_weights()))
model2.add(LeakyReLU(name='block4_relu'))
```

```python
########## optimize search ##########
from scipy.optimize import minimize
from sklearn.metrics import log_loss


best_acc = 0.0

best_weights = None

def log_loss_func(weights):
    ''' scipy minimize will pass the weights as a numpy array '''
    final_prediction = np.zeros((n_test, n_classes), dtype='float32')
    for weight, prediction in zip(weights, y_preds):
        final_prediction += weight * prediction

    return log_loss(y_private_onehot, final_prediction)

for iteration in range(n_ensembles):
    prediction_weights = np.random.random(n_ensembles)

    constraints = ({'type': 'eq', 'fun':lambda w: 1 - sum(w)})
    bounds = [(0, 1)] * n_ensembles

    result = minimize(log_loss_func, prediction_weights, method='SLSQP', bounds=bounds, constraints=constraints)

    print('Best Ensemble Weights: {weights}'.format(weights=result['x']))

    weights = result['x']
    weighted_predictions = np.zeros((n_test, n_classes), dtype='float32')

    for weight, prediction in zip(weights, y_preds):
        weighted_predictions += weight * prediction

    y_pred = np.argmax(weighted_predictions, axis=1)
    yTrue = y_private.reshape(-1)

    accuracy = metrics.accuracy_score(yTrue, y_pred) * 100
    error = 100 - accuracy
    print("Iteration %d: Accuracy : " % (iteration + 1), accuracy)
    print("Iteration %d: Error : " % (iteration + 1), error)

    if accuracy > best_acc:
        best_acc = accuracy
        best_weights = weights

    print()

print("Best Accuracy : ", best_acc)
print("Best Weights : ", best_weights)
```
